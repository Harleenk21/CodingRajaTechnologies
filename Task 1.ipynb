```python
import numpy as np
import pandas as pd
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia =SentimentIntensityAnalyzer()
```

```python
df=pd.read_csv("threads_reviews.csv")
df.head()
```

```python
df['review_description'].isnull().sum()
```

```python
df['scores']=df['review_description'].apply(lambda review_description :sia.polarity_scores(str(review_description)))
df.head()
```

```python
df['compound']=df['scores'].apply(lambda score_dict :score_dict['compound'])
df.head()
```

```python
df['pos']=df['scores'].apply(lambda pos_dict :pos_dict['pos'])
df.head()
```

```python

df['neg']=df['scores'].apply(lambda neg_dict :neg_dict['neg'])
df.head()
```

```python
df['t']=''
df.loc[df.compound>0,'t']='POS'
df.loc[df.compound==0,'t']='NEUTRAL'
df.loc[df.compound<0,'t']='NEG'
df.head()
```

```python
print(df.columns)
```

```python
print(df['t'].unique())
```

```python
df.head()
```

```python
rows = df.shape[0]
pos = 0
neg = 0
neutral = 0
for i in range(rows):
    t = df.loc[i, 't']
    if t == 'POS':
        pos = pos + 1
    elif t == 'NEG':
        neg = neg + 1
    elif t == 'NEUTRAL':
        neutral = neutral + 1
print("POSITIVE:", str(pos), "NEGATIVE:", str(neg), "NEUTRAL:", str(neutral))

```

```python
if(pos>neg and pos>neutral):
    print("positive reviews are more")
elif neg>neutral:
    print("negative reviews are more")    
else:
    print("neutral reviews are more")
          
```
