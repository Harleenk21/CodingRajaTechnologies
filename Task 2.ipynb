```python
x= df.iloc[:,0:13]  #diving data into x and y
x
```

```python
y=df.iloc[:,13]
y
```

```python
from sklearn.model_selection import train_test_split       #dividing data into train and test based on random state method
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.10,random_state=10)
```

```python
x_train  # but here the decimal values are discrete(1,4,7 etc)
```

```python
from sklearn.preprocessing import StandardScaler
scaler =StandardScaler()      # to call only standardize function and not to train(just call)
```

```python
scaler.fit_transform(x_train)   # fit is to give the standard scaler data to transform(standardize) and fit into x train data
```

```python
X_train =scaler.fit_transform(x_train) # only train
```

```python
X_test = scaler.transform(x_test) # only change 
```

```python
# to get data back (original data)
#scaler.inverse_transform(X_train)
```

```python
from sklearn.linear_model import LinearRegression
#cross validation
from sklearn.model_selection import cross_val_score
```

# Applying ML MODEL



```python
# estimator
regression = LinearRegression() 
regression.fit(X_train,y_train)
```

```python
cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=3)  #these values will be different for all systems as values are taken random so score will also vary 
```

```python
mse=cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=3) #storing values into mse
```

```python
import numpy as np
np.mean(mse)
```

```python
#prediction
reg_pred =regression.predict(X_test)
```

```python
reg_pred
```

```python
import seaborn as sns 
sns.displot(reg_pred-y_test)
```

# R Square

```python
from sklearn.metrics import r2_score
```

```python
score =r2_score(reg_pred,y_test)
```

```python
score
```

```python
# Plotting y_test and y_pred to understand the spread.
fig = plt.figure()
plt.scatter(y_test,reg_pred)
fig.suptitle('y_test vs reg_pred', fontsize=20)              # Plot heading 
plt.xlabel('y_test', fontsize=18)                          # X-label
plt.ylabel('reg_pred', fontsize=16)                          # Y-label
```
